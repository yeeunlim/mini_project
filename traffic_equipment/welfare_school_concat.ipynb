```python
import pandas as pd
```

```python
df_wel = pd.read_csv('./geocode_welfare_transformed.csv')
df_school = pd.read_csv('./서울시 특수학교 기본정보_변환.csv')
```

```python
df_wel = df_wel[['시설명', 'X', 'Y']]
df_wel.info()
```

```python
df_school.info()
```

```python
df_school = df_school.rename(columns={'경도': 'X', '위도': 'Y', '학교명': '시설명'})
df_school = df_school[['시설명', 'X', 'Y']]
df_school
```

```python
df_welfare = pd.concat([df_wel, df_school])
df_welfare = df_welfare.reset_index(drop=True)
df_welfare
```

```python
df_welfare.to_csv('welfare_school_concat.csv', index=False)
```
