```python
import pandas as pd
import folium

data = pd.read_excel('./보행자 사고다발지역.xlsx')
data
```

```python
data
```

```python
data['시도'] = data['시도'].str.replace('서울','')
data['시도'] = data['시도'].str.replace(r'\d+$', '', regex=True)
```

```python
data = data.iloc[:, [0,1,2,3,4]]
```

```python
print(data.isna().any().any())
```

```python
# 2022년 보행자 사고 다발지역
data
```

```python
seoul_map = folium.Map(location = [37.55, 126.98], zoom_start = 12)
seoul_map

for lat, lng, sido, place, acc_rate in zip(data['Latitude'], data['Longitude'], data['시도'], data['장소설명'], data['사고건수']):
    folium.Marker([lat, lng], popup = acc_rate).add_to(seoul_map)
```

```python
seoul_map
```

```python
map = folium.Map(location=[data['Latitude'].mean(), data['Longitude'].mean()], zoom_start=11)

# Function to determine color based on the number of accidents
def get_color(accidents):
    if accidents > 9:
        return 'red'
    elif accidents > 5:
        return 'orange'
    else:
        return 'green'

# Add markers to the map
for idx, row in data.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=row['사고건수'] * 2, # Dynamic radius based on the number of accidents
        popup=f"{row['장소설명']}, 사고건수: {row['사고건수']}",
        color=get_color(row['사고건수']),
        fill=True,
        fill_color=get_color(row['사고건수'])
    ).add_to(map)

```
