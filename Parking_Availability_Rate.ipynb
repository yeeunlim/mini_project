```python
import pandas as pd
```

```python
# 장애인 복지시설 데이터 불러오기
data = pd.read_excel('./total_accident_welfare_facilities.xlsx')
data
```

```python
# 고유한 '시군구명' 값들을 추출
unique_gu_names = data['시군구명'].unique()

# '시군구명'으로 데이터프레임을 정렬
data_sorted = data.sort_values(by='시군구명')

# 각 구의 데이터 개수를 구하는 데이터프레임 생성
gu_data_count = data_sorted['시군구명'].value_counts().reset_index()
gu_data_count.columns = ['자치구', '복지시설 개수']
```

```python
# 주차장 확보율 데이터 가져오기
data1 = pd.read_excel('./data/주차장+확보율_20240205103636.xlsx')

# 2022년 기준 데이터 추출
data1_op = data1.iloc[1:,[1, -1]]
```

```python
# 컬럼명 일치
data1_op.columns = ['자치구', '2022년 주차장확보울(%)']
data1_op
```

```python
# 정렬
data1_op = data1_op.sort_values(by='자치구').reset_index(drop=True)
```

```python
# 데이터 합치기
merge_df = pd.merge(data1_op, gu_data_count, on='자치구', how='left')
```

```python
merge_df = merge_df.drop(17).reset_index(drop=True)
```
