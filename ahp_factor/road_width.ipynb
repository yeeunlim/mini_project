```python
import pandas as pd
```

```python
# 파일 로드
df_welfare = pd.read_csv('../welfare_facillities_and_disabled_school/welfare_school_concat_EPSG_5186.csv')
df_road_width = pd.read_csv("./서울시_도로노선_정보.csv",encoding='euc-kr')
```

```python
df_welfare = df_welfare.head()
```

```python
desired_road = '백제고분로37길'
filtered_df_road_width = df_road_width[df_road_width['노선명(도로명)'].str.contains(desired_road)]
filtered_df_road_width
```

```python
# 결과를 저장할 데이터프레임 생성
df_welfare_road_width = pd.DataFrame(columns=['시설명', '시설주소', '도로규모'])

# '노선명(도로명)'이 '시설주소'에 포함되는지 확인하고 결과 저장
for _, row_welfare in df_welfare.iterrows():
    for _, row_road_width in df_road_width.iterrows():
        if row_road_width['노선명(도로명)'] in row_welfare['시설주소']:
            df_welfare_road_width = pd.concat([df_welfare_road_width, pd.DataFrame({
                '시설명': [row_welfare['시설명']],
                '시설주소': [row_welfare['시설주소']],
                '도로규모': [row_road_width['도로규모']]
            })], ignore_index=True)

# 결과 출력
print(df_welfare_road_width)
```
